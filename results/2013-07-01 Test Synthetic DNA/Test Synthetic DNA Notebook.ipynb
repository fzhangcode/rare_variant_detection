```python
def chisquaretest(pvector,X,lamda):
    nsum=np.sum(X)
    E=nsum*pvector
    X=np.array(X)

    if lamda==0 or lamda==-1:
        C=2.0/np.sum(X*np.log(X*1.0/E))
    else:
        C=2.0/(lamda*(lamda+1))*np.sum(X*((X*1.0/E)**lamda-1))
    df=len(pvector)-1
    #p=scipy.special.gammainc(C,df)
    p=1-gammainc(df/2,C/2)  
    return(p)
```
