```python
def movingaverage(interval, window_size):
    window=  np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'same')
```

```python
f = h5py.File('Case1_0.hdf5', 'r')
phi={}
phi['M'] = f['/phi/M'][...]
phi['M0'] = f['/phi/M0'][()]
loc = f['/loc'][...]
f.close()
```

```python
plt.semilogy(loc, phi['M'], color='blue', marker='+', linestyle='None', label='M')
plt.semilogy(loc, movingaverage(phi['M'],100), color='black', linewidth=2.0, label='Moving Average (100)' )
plt.xlabel('Location')
plt.ylabel('Replicate Precision')
legend()
```
